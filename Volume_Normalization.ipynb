```python
def create_contribution(df1, df2):
    
    df1 = df1[(df1['date']>'2020-05-31') & (df1['date']<'2022-07-01')].reset_index(drop = True)
    df1 = df1[~(((df1['platform_name']=='Flipkart')) & (df1['month'].isin(['2020-Oct', '2021-Oct'])))].reset_index(drop = True)
    print(df1.shape)
    df1 = df1.groupby(['asin', 'date']).sum().reset_index()
    df1['month'] = pd.to_datetime(df1['date']).dt.strftime("%Y-%b")
    
    temp1 = df1.groupby(['asin', 'month']).sum().reset_index()
    
    temp1 = temp1.rename(columns = {'vol_in_lit': 'vol_in_lit_d2m', 'vol_in_rum': 'vol_in_rum_d2m'})
        
    df2 = df2.rename(columns = {'vol_in_lit': 'vol_in_lit_monthly', 'vol_in_rum': 'vol_in_rum_monthly'})
    
    df2['month'] = pd.to_datetime(df2['month_date']).dt.strftime("%Y-%b")
      
    temp1 = pd.merge(df1, temp1, on = ['asin', 'month'])
    temp1['vol_in_lit_contrib'] = temp1['vol_in_lit']/temp1['vol_in_lit_d2m']
    temp1['vol_in_rum_contrib'] = temp1['vol_in_rum']/temp1['vol_in_rum_d2m']
        
    temp2 = df2.groupby(['asin', 'month']).sum().reset_index()
    print(temp2.columns)
    temp = pd.merge(temp1, temp2, left_on = ['asin', 'month'], right_on = ['asin', 'month'])
    
    temp['diff_vol_in_lit'] = temp['vol_in_lit_monthly'] - temp['vol_in_lit_d2m']
    temp['diff_vol_in_rum'] = temp['vol_in_rum_monthly'] - temp['vol_in_rum_d2m']
    
    temp['adj_vol_in_lit'] = temp['vol_in_lit'] + temp['diff_vol_in_lit']*temp['vol_in_lit_contrib']
    temp['adj_vol_in_rum'] = temp['vol_in_rum'] + temp['diff_vol_in_rum']*temp['vol_in_rum_contrib']
    
    return temp[['date', 'month', 'asin', 'vol_in_lit', 'vol_in_rum',
                           'adj_vol_in_lit', 'adj_vol_in_rum']]

final_df = create_contribution(vol_daily[['date', 'month', 'platform_name', 'asin', 'vol_in_lit', 'vol_in_rum']],
                   vol_monthly[['month_date', 'month', 'asin', 'vol_in_lit', 'vol_in_rum']])
final_df
```

```python
def add_bb_volume(vol,df,col_name,start_date,end_date):
    
    start = df[df['date']==start_date].index[0]
    end = df[df['date']==end_date].index[0]+1
    no_of_days = end - start

    new_df = df.copy()
    new_df['FLIPKART_VOL1'] = 0; new_df[['FVOL']]=0
    sdf = pd.DataFrame(columns=['date','VOL','MA'])
    for i,r in new_df[[col_name]].iloc[start:end].iterrows():
        new_df['FLIPKART_VOL1'].iloc[i] = round(vol/no_of_days,2)
    new_df['FVOL'] = round(new_df[col_name] + new_df['FLIPKART_VOL1'],2)

    for i,r in new_df[['FVOL']].iloc[start:end].iterrows():
        d = new_df['date'].iloc[i]
        a = round(new_df['FVOL'].iloc[i],2)
        b = round(new_df['FVOL'].iloc[i-3:i+4].mean(),2)
        t1 = pd.DataFrame([[d,a,b]], columns=['date','VOL','MA'])
        sdf = pd.concat([sdf,t1])
    sdf['adj_'+ col_name] = round((sdf['VOL'].sum() - sdf['MA'].sum())/no_of_days,2) + sdf['MA']

    return sdf[['date', 'adj_'+ col_name]]
```

```python
def big_billion_volumes(df, monthly_df):
    
    df = df[(df['platform_name']=='Flipkart') & (df['month'].isin(['2020-Oct']))][['date', 'month', 'asin', 'vol_in_lit', 'vol_in_rum']]
    sku_list = df['asin'].unique()
    adj_df = pd.DataFrame()
    for i in sku_list:
        date_df = pd.DataFrame(columns = ['date', 'asin'])
        date_df['date'] = pd.date_range('2020-10-01', '2020-10-31',freq='d')
        date_df['asin'] = i
        new_df = df[df['asin']==i].reset_index(drop = True)
        temp_df = new_df.copy()
        temp_df = pd.merge(temp_df, date_df, on = ['asin', 'date'], how = 'outer')
        temp_df = temp_df.sort_values(by = 'date').reset_index(drop = True)

        monthly_vol_in_lit = monthly_df[(monthly_df['asin']==i) & \
                                (monthly_df['month'].isin(['2020-Oct']))].groupby('asin').sum().reset_index()['vol_in_lit'].values[0]
        monthly_vol_in_rum = monthly_df[(monthly_df['asin']==i) & \
                                (monthly_df['month'].isin(['2020-Oct']))].groupby('asin').sum().reset_index()['vol_in_rum'].values[0]

        bb_vol_in_lit = monthly_vol_in_lit - temp_df.groupby('month').sum()['vol_in_lit'].values[0]
        bb_vol_in_rum = monthly_vol_in_rum - temp_df.groupby('month').sum()['vol_in_rum'].values[0]

        start_date = '2020-10-15'; end_date = '2020-10-21'
        
        bb_df_vol_lit = add_bb_volume(bb_vol_in_lit,temp_df[['date', 'asin', 'vol_in_lit']].fillna(0), 'vol_in_lit',start_date,end_date)
        bb_df_vol_rum = add_bb_volume(bb_vol_in_rum,temp_df[['date', 'asin', 'vol_in_rum']].fillna(0), 'vol_in_rum',start_date,end_date)
        bb_df_vol_lit['asin'] = i; bb_df_vol_rum['asin'] = i

        final_df = pd.merge(pd.concat([new_df, bb_df_vol_lit]), bb_df_vol_rum, on = ['asin', 'date'], how = 'outer')
        final_df['month'] = final_df['month'].fillna(method="ffill").fillna(method='bfill')
        final_df = final_df.fillna(0)
        final_df['adj_vol_in_lit'] = final_df['vol_in_lit'] + final_df['adj_vol_in_lit']
        final_df['adj_vol_in_rum'] = final_df['vol_in_rum'] + final_df['adj_vol_in_rum']
        final_df = final_df.sort_values('date').reset_index(drop = True)
        
        adj_df = pd.concat([adj_df, final_df], axis=0)
        
    return adj_df
    
        
bb_df_2020 = big_billion_volumes(vol_daily, vol_monthly)
bb_df_2020
```

```python
def big_billion_2021(df, monthly_df):
    df = df[df['month'].isin(['2021-Oct'])]
    amz_df = df[(df['platform_name']=='Amazon') & (df['date'].between('2021-10-02', '2021-10-31'))]
    contrib_df = pd.DataFrame()
    contrib_df = amz_df.groupby('date').sum()[['vol_in_lit', 'vol_in_rum']].reset_index()
    contrib_df['vol_in_lit_contrib'] = contrib_df['vol_in_lit']/sum(contrib_df['vol_in_lit'])
    contrib_df['vol_in_rum_contrib'] = contrib_df['vol_in_rum']/sum(contrib_df['vol_in_rum'])
#     print(contrib_df)
    
    df = df[(df['platform_name']=='Flipkart') & (df['month'].isin(['2021-Oct']))][['date', 'month', 'asin', 'vol_in_lit', 'vol_in_rum']]
    sku_list = df['asin'].unique()
    
    adj_df = pd.DataFrame()
    for i in sku_list:
        date_df = pd.DataFrame(columns = ['date', 'asin'])
        date_df['date'] = pd.date_range('2021-10-01', '2021-10-31',freq='d')
        date_df['asin'] = i
        new_df = df[df['asin']==i].reset_index(drop = True)
        temp_df = new_df.copy()
        temp_df = pd.merge(temp_df, date_df, on = ['asin', 'date'], how = 'outer')
        temp_df = temp_df.sort_values(by = 'date').reset_index(drop = True)

        monthly_vol_in_lit = monthly_df[(monthly_df['asin']==i) & \
                                (monthly_df['month'].isin(['2021-Oct']))].groupby('asin').sum().reset_index()['vol_in_lit'].values[0]
        monthly_vol_in_rum = monthly_df[(monthly_df['asin']==i) & \
                                (monthly_df['month'].isin(['2021-Oct']))].groupby('asin').sum().reset_index()['vol_in_rum'].values[0]

        bb_vol_in_lit = monthly_vol_in_lit - temp_df.groupby('month').sum()['vol_in_lit'].values[0]
        bb_vol_in_rum = monthly_vol_in_rum - temp_df.groupby('month').sum()['vol_in_rum'].values[0]
        
        temp_df = pd.merge(temp_df, contrib_df[['date', 'vol_in_lit_contrib', 'vol_in_rum_contrib']], on = 'date', how = 'outer')

        temp_df['month'] = temp_df['month'].fillna(method="ffill").fillna(method='bfill')
        temp_df['adj_vol_in_lit'] = temp_df['vol_in_lit'].fillna(bb_vol_in_lit * temp_df['vol_in_lit_contrib'])
        temp_df['adj_vol_in_rum'] = temp_df['vol_in_rum'].fillna(bb_vol_in_rum * temp_df['vol_in_rum_contrib'])
        
        temp_df = temp_df.fillna(0)
       
        adj_df = pd.concat([adj_df, temp_df[['date', 'month', 'asin', 'vol_in_lit', 'vol_in_rum',
                           'adj_vol_in_lit', 'adj_vol_in_rum']]], axis=0)
        
    return adj_df
    
bb_df_2021 = big_billion_2021(vol_daily, vol_monthly)
bb_df_2021
```
